```python
# 1️⃣ Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2️⃣ Separate features and target
X = dfp.drop(columns=['name', 'status'])
y = dfp['status']

# 3️⃣ Balance the dataset using random undersampling
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)

print("Class distribution after undersampling:")
print(pd.Series(y_res).value_counts())

# 4️⃣ Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

# 5️⃣ Feature scaling (important for KNN)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# 6️⃣ Train KNN Classifier
knn = KNeighborsClassifier(n_neighbors=5)  # you can tune n_neighbors
knn.fit(X_train_scaled, y_train)

# 7️⃣ Predictions
y_pred = knn.predict(X_test_scaled)

# 8️⃣ Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# 1️⃣ Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2️⃣ Separate features and target
X = df_kid_filled.drop(columns=['id', 'age', 'classification'])
y = df_kid_filled['classification']

# 3️⃣ Balance the dataset using RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)

print("Class distribution after undersampling:")
print(pd.Series(y_res).value_counts())

# 4️⃣ Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

# 5️⃣ Initialize Decision Tree Classifier
dt = DecisionTreeClassifier(random_state=42)

# 6️⃣ Train the model
dt.fit(X_train, y_train)

# 7️⃣ Make predictions
y_pred = dt.predict(X_test)

# 8️⃣ Evaluate the model
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))
print("\n🔎 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
# 1️⃣ Import libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from imblearn.under_sampling import RandomUnderSampler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# 2️⃣ Separate features and target
X = df_lvr.drop(columns=['Age', 'Gender', 'Dataset'])
y = df_lvr['Dataset']

# 3️⃣ Balance the dataset using RandomUnderSampler
rus = RandomUnderSampler(random_state=42)
X_res, y_res = rus.fit_resample(X, y)

# 4️⃣ Split into train and test sets
X_train, X_test, y_train, y_test = train_test_split(
    X_res, y_res, test_size=0.2, random_state=42, stratify=y_res
)

# 5️⃣ Initialize Random Forest Classifier
rf = RandomForestClassifier(random_state=42, n_estimators=100)

# 6️⃣ Train the model
rf.fit(X_train, y_train)

# 7️⃣ Make predictions
y_pred = rf.predict(X_test)

# 8️⃣ Evaluate performance
print("✅ Accuracy:", accuracy_score(y_test, y_pred))
print("\n📊 Classification Report:\n", classification_report(y_test, y_pred))
print("\n🔍 Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```
